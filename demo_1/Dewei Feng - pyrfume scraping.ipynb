```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# 1) Set up headless Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")        # Run Chrome headless (no GUI)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--window-size=1920,1080")  # Important for headless

ingredients = {}

for food, food_id in food_ids.items():
    driver = webdriver.Chrome(options=chrome_options)

    try:
        # 2) Navigate to the page
        url = f"https://foodb.ca/foods/{food_id}"
        print(f"Navigating to: {url}")
        driver.get(url)

        # 3) Wait for page to load (increased wait time)
        time.sleep(12)  # Increased from 10 to 12 seconds

        # 4) Find the compounds table
        tables = driver.find_elements(By.CSS_SELECTOR, "table.display.food-contents")
        if not tables:
            print("No tables found. The page structure might have changed.")
            continue  # Skip to next food item if no tables found

        compounds_table = tables[0]

        # 5) Get rows in the <tbody>
        rows = compounds_table.find_elements(By.CSS_SELECTOR, "tbody tr")
        print(f"Total compound rows found: {len(rows)}")

        # 6) Extract the top 20 rows
        top_20 = rows[:20]
        compound_info_list = []

        for row in top_20:
            # Try getting text in two different ways
            try:
                cells = row.find_elements(By.TAG_NAME, "td")
                if len(cells) >= 4:
                    # First try with direct text property
                    compound_name = cells[0].text.strip()
                    average_str = cells[3].text.strip()

                    # If empty, try with textContent
                    if not compound_name:
                        compound_name = cells[0].get_attribute('textContent').strip()
                        average_str = cells[3].get_attribute('textContent').strip()

                    if compound_name:  # Only add if we got content
                        compound_info_list.append({
                            "compound_name": compound_name,
                            "average": average_str
                        })
            except Exception as e:
                print(f"Error processing row: {e}")
                continue

        print("\n=== Top Compounds ===")
        for i, c in enumerate(compound_info_list[:10], start=1):
            print(f"{i}. {c}")

        ingredients[food] = compound_info_list

    except Exception as e:
        print(f"Error processing {food}: {str(e)}")
        continue

    finally:
        driver.quit()

# Print final results
print("\nFinal Results:")
for food, compounds in ingredients.items():
    print(f"\n{food}:")
    for i, compound in enumerate(compounds[:10], 1):
        print(f"  {i}. {compound}")
```

```python
import re
import pandas as pd
import numpy as np
from chemparse import parse_formula

def parse_mg_per_100g(value_str):
    """
    Given a string like '21719.527 mg/100 g' or '4788.750 mg/100 g',
    return the numerical value as a float (e.g., 21719.527, 4788.75).

    If the string doesn't match or is missing, return 0.
    """
    if not value_str:
        return 0.0

    # Attempt to extract the leading numeric portion before " mg/100 g"
    # We'll allow for possible patterns like "7601.500", "3500.000 mg/100 g", or partial mg/100 g.
    match = re.search(r"([\d\.]+)", value_str)
    if match:
        return float(match.group(1))
    return 0.0

def build_food_embeddings_with_mass(data, agg_func='sum'):
    """
    Given the data structure:
       {
         'food_name': [
            {
              'compound_name': 'Water',
              'average': '21719.527 mg/100 g',
              'formula': 'H2O'
            },
            ...
         ],
         ...
       }

    1) Parse each compound's formula with chemparse (when valid).
    2) Convert the 'average' string to a float (mg/100 g).
    3) Multiply each element's count by the compound's mg value => "element_score".
    4) Aggregate (sum or mean) these element scores for each food => single embedding per food.

    Returns a pandas DataFrame with one row per food and columns for each element.
    """
    all_elements = set()
    rows = []

    # 1) Gather data in a "long" form: (food, compound, {element scores})
    for food_name, compounds in data.items():
        for formula, comp in compounds.items():

            formula = comp.get("formula", "")
            avg_str = comp.get("average", "")

            # Convert average mg/100 g to float
            mg_value = parse_mg_per_100g(avg_str)

            # Try to parse formula if it's not "N/A"
            if formula.startswith("N/A"):
                # We'll skip or treat as zero
                parsed = {}
            else:
                try:
                    parsed = parse_formula(formula)  # e.g. {'C': 6, 'H': 12, 'O': 6}
                except:
                    parsed = {}

            # Multiply each element count by mg_value to get a "weighted" count
            element_scores = {}
            for el, count in parsed.items():
                element_scores[el] = count * mg_value

            # Keep track of which elements exist globally
            all_elements.update(element_scores.keys())

            # Store row
            row_data = {
                "food_name": food_name,
                "compound_name": comp.get("compound_name", ""),
            }
            row_data.update(element_scores)  # add the per-element weighted counts
            rows.append(row_data)

    # 2) Create a long DataFrame with columns = [food_name, compound_name, <elements>...]
    df_long = pd.DataFrame(rows)

    # Fill missing element columns with 0
    for el in all_elements:
        if el not in df_long.columns:
            df_long[el] = 0.0

    # 3) Aggregate by food_name
    element_cols = sorted(list(all_elements))

    if agg_func == 'sum':
        df_agg = df_long.groupby("food_name")[element_cols].sum()
    elif agg_func == 'mean':
        df_agg = df_long.groupby("food_name")[element_cols].mean()
    else:
        raise ValueError("agg_func must be 'sum' or 'mean'")

    # 4) Return a DataFrame: one row per food, columns are elements
    df_agg = df_agg.reset_index()
    return df_agg
```
